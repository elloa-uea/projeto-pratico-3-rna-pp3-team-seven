```python
def statistics(arr):
    mean = np.mean(arr)
    median = np.median(arr)
    var = np.var(arr)
    std = np.std(arr)
    
    result = [round(mean, 2), round(median, 2), round(var, 2), round(std, 2)]
    
    text = "Average: "+str(result[0])+"\nMedian: "+str(result[1])+"\nVariance: "+str(result[2])+"\nStandard deviation:"+str(result[3])
    
    return text
plt.rcParams["figure.figsize"] = [10,9]
```

```python
# Number of occurrences for each 'Elevation' attribute value
new_data = dados['Elevation'].value_counts()

# Exclusive 'Elevation' dataframe
new_data_df = list(zip(new_data.index, new_data))
new_data_df = pd.DataFrame(new_data_df, columns = ['Elevation', 'Number of occurrences'])

# Graphic
ax = new_data_df.plot.scatter(x='Elevation',y='Number of occurrences', s=2, colormap='viridis')
ax.set_title('Elevation / Number of occurrences')
props = dict(boxstyle='Sawtooth', facecolor='white', alpha=0.3, pad=.5)
ax.text(1.1, 1, statistics(new_data_df['Number of occurrences']), transform=ax.transAxes, fontsize=12, verticalalignment='top', bbox=props)
```
