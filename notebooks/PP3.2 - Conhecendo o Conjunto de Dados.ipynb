```python
# 1. Baseando-se nos fundamentos de visualiza√ß√£o de dados abordados na disciplina, plote cada um
# dos atributos preditores de maneira a enfatizar a sua distribui√ß√£o, tend√™ncia central e dispers√£o 

plt.rcParams["figure.figsize"] = [6,5]

def graph_generator_soil_type_or_wilderness_area(attribute, len_types, point_size=2):
    array_attribute = []
    array_occurrences = []
    
    for i in range(1,len_types+1):
        occurrences = len(dados.loc[dados[attribute+str(i)] == 1])
        
        array_attribute.append(i)
        array_occurrences.append(occurrences)

        aux = list(zip(array_attribute, array_occurrences))
        
        # Dataframe with the occurrences and number of occurrences of the 'Soil_Type' or 'Wilderness_Area' attribute
        df_attribute = pd.DataFrame(aux, columns = [attribute, 'Number of occurrences'])

    new_data_df = df_attribute
    fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,6))
    # Graphic 1
    ax = new_data_df.plot.scatter(ax = axes[0],x=attribute,y='Number of occurrences', s=point_size, colormap='viridis')
    ax.set_title(attribute+' / Number of occurrences')
    # Graphic 2
    ax = new_data_df.boxplot()
    ax.set_title(attribute+' / Number of occurrences')
    props = dict(boxstyle='Sawtooth', facecolor='white', alpha=0.3, pad=.5)
    ax.text(1.1, 1, statistics(new_data_df['Number of occurrences']), transform=ax.transAxes, fontsize=12, verticalalignment='top', bbox=props)
    

def statistics(arr):
    mean = np.mean(arr)
    median = np.median(arr)
    var = np.var(arr)
    std = np.std(arr)
    result = [round(mean, 2), round(median, 2), round(var, 2), round(std, 2)]
    
    # Text with statistics values
    text = "     Number of occurrences\n"
    text = text + "\nAverage: "+str(result[0])+"\nMedian: "+str(result[1])+"\nVariance: "+str(result[2])+"\nStandard deviation:"+str(result[3])
    return text
    
def graph_generator(attribute, point_size=2):
    # Number of occurrences for each 'Attribute' attribute value
    new_data = dados[attribute].value_counts()

    # Exclusive 'Attribute' dataframe
    new_data_df = list(zip(new_data.index, new_data))
    new_data_df = pd.DataFrame(new_data_df, columns = [attribute, 'Number of occurrences'])

    fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,6))
    # Graphic 1
    ax = new_data_df.plot.scatter(ax = axes[0],x=attribute,y='Number of occurrences', s=point_size, colormap='viridis')
    ax.set_title(attribute+' / Number of occurrences')
    # Graphic 2
    ax = new_data_df.boxplot()
    ax.set_title(attribute+' / Number of occurrences')
    props = dict(boxstyle='Sawtooth', facecolor='white', alpha=0.3, pad=.5)
    ax.text(1.1, 1, statistics(new_data_df['Number of occurrences']), transform=ax.transAxes, fontsize=12, verticalalignment='top', bbox=props)
    
graph_generator('Elevation')
graph_generator('Aspect')
graph_generator('Slope')
graph_generator('Horizontal_Distance_To_Hydrology')
graph_generator('Vertical_Distance_To_Hydrology')
graph_generator('Horizontal_Distance_To_Roadways')
graph_generator('Hillshade_9am')
graph_generator('Hillshade_Noon')
graph_generator('Hillshade_3pm')
graph_generator('Horizontal_Distance_To_Fire_Points')

graph_generator_soil_type_or_wilderness_area('Soil_Type',40,40)      # ---> graph generator of 'Soil_Type'
graph_generator_soil_type_or_wilderness_area('Wilderness_Area',4,40) # ---> graph generator of 'Wilderness_Area1'
```
